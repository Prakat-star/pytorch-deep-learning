### Matrix Multiplication

Two main ways of performing multiplication in neural networks and deep learning

1. Element-wise Matrix Multiplication
2. Matrix Multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimentions** must match
2. The resulting matric has the shape of the **outer dimentions**


```python
# Element-wise multiplication
print(tensor ,"*", tensor)
print(f"Equals : {tensor * tensor}")
```

```python
# Matrix Multiplication
torch.matmul(tensor,tensor)
```

```python
# Matrix Multiplication by hand
value = 0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
print(value)
```

### One of the most common erroros in deep learning : shape errors


```python
# shapes for matrix multiplication

tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

# torch.matmul(tensor_A , tensor_B)
torch.mm(tensor_A , tensor_B) # torch.mm is the same as torch.matmul (it;s an alias for matmul)

# it doesn;t work cus the shapes for both matrix is (3,2) for both , this shape can;e be multiplied 
```

```python
tensor_A.shape , tensor_B.shape
```

To fix out tensor shape issue , we can manipulate the shape of one of our tensor using a **transpose**

```python
tensor_B
```

```python
tensor_B.T
```

```python
# The matrix multiplication operation works when tensor_B is transposed

torch.mm(tensor_A , tensor_B.T)
```

```python
print(f"Original shapes : tensor_A = {tensor_A.shape} , tensor_B = {tensor_B.shape}")
print(f"New shapes : tensor_A = {tensor_A.shape} , tensor_B = {tensor_B.T.shape} ")
print(f"Multiplying : {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimentions must match")
print("Output: \n")
output = torch.mm( tensor_A , tensor_B.T)
print(output)
print(f"\n Output Shape : {output.shape}")
```

### Reshaping , Stacking , squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an impot tensor of a certain shape but keep the same memory
* Stacking - combining multiple tensors on top of each other (vstack) , or side by side (hstack).
* Squeeze - removes all `1` dimentions from a tensor
* unsqueexe - add a `1` dimention to a target tensor
* permute - return a view of the input with dimentions permutted (swapped) in a certain way




```python
# creating a tensor

import torch
x = torch.arange(1.,11.)
x , x.shape
```

```python
# Add an extra dimensions
x_reshaped = x.reshape(1 , 10)
x_reshaped , x_reshaped.shape
```

```python
x_reshaped = x.reshape(10 , 1)
x_reshaped , x_reshaped.shape
```

```python
x_reshaped = x.reshape(5 , 2)
x_reshaped , x_reshaped.shape

# and so on , as long as the reshape is compatible with the original shape
```

```python
# Change the view
z = x.view( 1 , 10 )
z , z.shape
```

```python
#changing z changed x [because a view of a tensor shares the same memory as the original]

z[ : , 0] = 5
z , x
```

```python
# stack tensors on top of each other

x_stacked = torch.stack([x,x,x,x] , dim=0)
x_stacked , x_stacked.shape

# vstack is dim = 0 and hstack is dim = 1
```

```python
x = torch.arange(1.,10.)
x_reshaped = x.reshape(1 , 9)

print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
#torch.unsqueeze() - adds a single dimention to a target tensor at a specific dim (dimention)

print(f"Previous targer :  {x_squeezed}")
print(f"Previous shape : {x_squeezed.shape}")

#Add an extra dimention with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges  the dimentions of a target tensor in a specified order

x_original = torch.rand(size=(224,224,3)) #height , width , color channels

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute( 2 , 0 , 1 ) #shifts axis 0->1 , 1->2 and 2->0

print(f"Previous shape : {x_original.shape}")
print(f"New shape : {x_permuted}")
```

```python
x_original[0,0,0] = 67
x_original , x_permuted
```
