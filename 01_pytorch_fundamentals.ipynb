### Matrix Multiplication

Two main ways of performing multiplication in neural networks and deep learning

1. Element-wise Matrix Multiplication
2. Matrix Multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimentions** must match
2. The resulting matric has the shape of the **outer dimentions**


```python
# Element-wise multiplication
print(tensor ,"*", tensor)
print(f"Equals : {tensor * tensor}")
```

```python
# Matrix Multiplication
torch.matmul(tensor,tensor)
```

```python
# Matrix Multiplication by hand
value = 0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
print(value)
```

### One of the most common erroros in deep learning : shape errors


```python
# shapes for matrix multiplication

tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

# torch.matmul(tensor_A , tensor_B)
torch.mm(tensor_A , tensor_B) # torch.mm is the same as torch.matmul (it;s an alias for matmul)

# it doesn;t work cus the shapes for both matrix is (3,2) for both , this shape can;e be multiplied 
```

```python
tensor_A.shape , tensor_B.shape
```

To fix out tensor shape issue , we can manipulate the shape of one of our tensor using a **transpose**

```python
tensor_B
```

```python
tensor_B.T
```

```python
# The matrix multiplication operation works when tensor_B is transposed

torch.mm(tensor_A , tensor_B.T)
```

```python
print(f"Original shapes : tensor_A = {tensor_A.shape} , tensor_B = {tensor_B.shape}")
print(f"New shapes : tensor_A = {tensor_A.shape} , tensor_B = {tensor_B.T.shape} ")
print(f"Multiplying : {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimentions must match")
print("Output: \n")
output = torch.mm( tensor_A , tensor_B.T)
print(output)
print(f"\n Output Shape : {output.shape}")
```
